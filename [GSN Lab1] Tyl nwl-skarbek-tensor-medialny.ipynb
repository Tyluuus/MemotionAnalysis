#Reprezentacje tensorowe obrazu i dźwięku

**Tensor sygnałowy** $T$ to matematyczna reprezentacja sygnałowego obiektu cyfrowego. Opiera się ona na określeniu: 
- Dziedziny $I$ tensora jako zbioru wielo-indeksów $[i_0,\dots,i_{d-1}]$ o wymiarze $d\geq 1$, w których każdy indeks składowy $i_k$ może należeć do zbioru o indywidulanym rozmiarze $n_k$, np. $i_k\in[0,n_k)$. Pierwszy z tych indeksów może np. numerować obiekt w kolekcji analizowanych obiektów cyfrowych, jeden z tych indeksów przeznaczony jest na numerację mierzonych lub obliczanych cech (w tym tzw. cech głębokich generowanych w sieciach neuronowych), a pozostałe specyfikują dyskretne współrzędne przestrzenne i/lub czasowe elementu tensora.
- Przypisania każdemu indeksowi $[i_0,\dots,i_{d-1}]$ wartości liczbowej $T[i_0,\dots,i_{d-1}]\in\bb{R}$, która reprezentuje wartość mierzonej cechy obiektu (np. poziom czerwieniu w obrazie kamerowym) lub wartość obliczaną (np. intensywność cechy głębokiej o numerze $i_1$ w mapie cech obliczanej przez jednostkę $U$ pewnej sieci neuronowej trenowanej w rozpoznawaniu twarzy w obrazie).

***Jak reprezentujemy tensory w Pythonie?***

Mozna to robić na wiele sposobów. Docelowo będziemy stosować tablice pakietu `NumPy` (typ `ndarray`), ale tutaj dla utrwalenia powyższych definicji zastosujmy listową strukturę danych Pythona, która jest klasą wbudowaną w język skryptowy Python.

###Przykład tensora 1D reprezentujący sygnał harmoniczny

Importujemy pakiet `math` z elementarnymi funkcjami i pakiet `matplotlib.pyplot` stosowany w sporządzaniu wykresów:

````
import math
import matplotlib.pyplot as plt
````

Określamy parametry tensora sygnałowego oraz listę indeksów chwil czasowych, a więc dziedzinę tensora:

````
N = 200; I = [i for i in range(N)] # indeks tensora
K = 197 # sprawdź K = 7, 77, 117, 177, 197, 199, 200
````

Wyznaczamy wartości próbek sygnału na zdefiniowanej indeksowej dziedzinie `I`:

````
T = [math.sin(math.pi*i*K/N+.1) for i in I] # tensor jako przypisanie
````

*Sporządzamy wykres*:
- utożsamiamy indeksy próbek ze współrzędnymi $x$,
- stosujemy procedurę rysowania linii łamanych `plot`: 
  - odcinek łamanej prowadzi od punktu $(i,T[i])$ do punktu $(i+1,T[i+1])$,
  - pierwszy argument w `plot`: lista współrzędnych $x$ dla punktów łamanej,
  - drugi argument w `plot`: lista współrzędnych $y$ dla punktów łamanej, 
  - opcjonalny trzeci argument określa tutaj kolor łamanej.

````
plt.plot(I,T); plt.plot([0,N-1],[0,0],'r'); plt.plot([0,0],[-1,1],'r'); plt.show() 
# niebieski wykres sygnału T, czerwona oś czasu
````

*Mini ćwiczenie*: Nie zawsze łamana musi określać przybliżenie jakiejś funkcji. Np. możesz narysować czerwoną oś pionową rysując odcinek od punktu $(0,1)$ do punktu $(0,-1)$ $-$, a więc spróbuj dopisać dodatkową linijkę do powyższej komórki kodowej.

*Krytyka powyższego kodu tworzenia szablonu szachownicy*:
- szachownica tworzona jest przez generator listowy,
- generator listowy dopisuje kolejne elementy w pętli,
- pętla w interpreterze jest kosztowna w porównaniu do tworzenia list przez funkcje wbudowane, które są realizowane w C/C++.


*Przykład przyspieszenia generowania szablonu szachownicy, gdy stosujemy listowe funkcje wbudowane zamiast listowych generatorów*:

````
def chessA(n): 
  I = [i for i in range(n)]
  return [[(1+(-1)**(i+k))//2 for i in I] for k in I]
def chessB(n):
  k = n//2; z = [0,1]*k; o = [1,0]*k
  if n&1: z.append(0); o.append(1) 
  oz = [o,z]*k
  if n&1: oz.append(o) 
  return oz
````

````
from time import clock
````

````
n = 1001
t = clock(); cA = chessA(n); dtA = clock()-t
print('czas A:', dtA, '[s]')
t = clock(); cB = chessB(n); dtB = clock()-t
print('czas B:', dtB, '[s]')
print('przyspieszenie:', dtA/dtB)
````

Dobrym kandydatem do utworzenia takiej funkcji jest `wytnij`. Nie tylko zajmuje więcej miejsca, ale też w rozwiązaniu będzie używana dwa razy.

````
def wytnij(RGB,x,y,w,h): 
  '''
  wycina w obrazie RGB  okno 2w+1 na 2h+1 o środku w (x,y)
  '''
  H,W,c = RGB.shape 
  xmin = max(x-w,0); xmax = min(x+w+1,W)
  ymin = max(y-h,0); ymax = min(y+h+1,H)
  rgb = RGB[ymin:ymax,xmin:xmax,:]
  return rgb
````

To samo dotyczy dwukrotnej realizacji punktów 1-5 w naszym zadaniu $-$ tak, raz dla koloru najbliższego bieli i raz dla najbliższego czerni. Zauważ numeracje kroków w komentarzach w linii rozpoczynającej wykonanie tego kroku.

````
def atask(T,x,y,w,h):
  Tp = wytnij(T,x,y,w,h) # 1.
  plt.imsave('lena-white.jpg',Tp) # 2.
  fig, (ax1, ax2) = plt.subplots(nrows=1,ncols=2) # 3.
  fig.set_size_inches(12,24); ax1.imshow(T); ax2.imshow(Tp); plt.show()
  R,G,B = T[:,:,0],T[:,:,1],T[:,:,2]; r,g,b = Tp[:,:,0],Tp[:,:,1],Tp[:,:,2]
  print('średni kolor w obrazie RGB:',int(.5+np.mean(R)), # 4.
        int(.5+np.mean(G)),int(.5+np.mean(B)))
  print('średni kolor w obrazie rgb:',int(.5+np.mean(r)),
        int(.5+np.mean(b)),int(.5+np.mean(b)))
  Y = (0.299*R+0.587*G+0.114*B).astype('uint8') # 5.
  print('średnia luminancja w obrazie Y:',int(.5+np.mean(Y)))
  y = (0.299*r+0.587*g+0.114*b).astype('uint8')
  print('średnia luminancja w obrazie y:',int(.5+np.mean(y)))
````

Tu zaczyna sie wykonanie naszej aplikacji, która w całości stanowi rozwiązanie zadania. Najpierw więc tworzymy tensor obrazu i ustalamy jego rozmiary: 

````
T = plt.imread('lena.jpg')
h,w,c = T.shape; print('wysokość:',h,'\nszerokość:',w,'\ngłębokość:',c)
````

Teraz tworzymy 2D tensory $R, G, B, R', G' ,B', S$, w którym:
- $R, G, B$ to odseparowane składowe obrazu $RGB$, 
- $R' ,G', B'$ to składowe odniesione do punktu bieli, a 
- $S{yx}\eqd (R')^2_{yx}+(G')^2_{yx}+(B')^2_{yx}$ to kwadraty odległości kolorów od punktu bieli. 

Poprawność tych wszystkich operacji arytmetycznych wymaga większego zakresu niż daje  wejściowy typ danych o nazwie 'unint8`, a więc przedział dyskretny $[0,255]$. Dlatego zaczynamy od konwersji wejściowego tensora $T$ do typu `int32` o dużo szerszym zakresie $[-2^{31},2^{31}).$

````
T32 = T.astype('int32'); R,G,B = T32[:,:,0],T32[:,:,1],T32[:,:,2]
Rp = R-255; Gp = G-255; Bp = B-255; S = Rp*Rp+Gp*Gp+Bp*Bp
print('min squared norm for pixel color:',np.min(S))
````

Równoważny, bardziej efektywny kod, nie wymaga separowania składowych, bowiem operacja sumowania elementów tensora wzdłuż wybranej osi  jego układu współrzędnych pozwala bezpośrednio policzyć sumę kwadratów odległości od punktu bieli.


````
T32 = T.astype('uint32')-255; S = np.sum(T32*T32,axis=-1)
print('min squared distance (color space) from white point:',np.min(S))
````

Kolejna linijka kodu, pozwoli znaleźć najpierw indeks miejsce w buforze tensora $S$ o najmniejszej wartości eleemntu (funkcja `argmin`), a potem zamienić ten indeks na współrzędne przestrzenne $y,x$ (funkcja `unravel_index`).

````
i = np.argmin(S,axis=None); y,x = np.unravel_index(i,S.shape)
i,y,x,S[y,x]
````
